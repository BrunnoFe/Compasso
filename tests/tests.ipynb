```python
import tkinter as tk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure
import random

# Initial setup
root = tk.Tk()
fig = Figure(figsize=(5, 4))
ax = fig.add_subplot(111)
line, = ax.plot([], [], 'r-')
canvas = FigureCanvasTkAgg(fig, master=root)
canvas.get_tk_widget().pack()

x_data, y_data = [], []

def update_plot():
    # Append data, update line object, and redraw
    y_data.append(random.randint(0, 100))
    x_data.append(len(y_data))
    line.set_data(x_data, y_data)
    ax.relim()
    ax.autoscale_view()
    canvas.draw()
    root.after(100, update_plot) # Repeat every 100ms

update_plot()
root.mainloop()
```

```python
import os
import pandas as pd

# Define your target folder and output file path
folder_path = r"C:\Users\shaulimdoplim\Documents\GitHub\Compasso\musicas"
output_excel = "file_list.xlsx"

# 1. Collect all file names and paths
file_data = []
for root, dirs, files in os.walk(folder_path):
    for file in files:
        full_path = os.path.join(root, file)
        file_data.append({
            "File Name": file,
            "Folder Location": root,
            "Full Path": full_path
        })

# 2. Convert to a DataFrame and export to Excel
df = pd.DataFrame(file_data)
df.to_excel(output_excel, index=False)

print(f"Successfully saved {len(df)} files to '{output_excel}'.")
```

```python
import platform
import subprocess

def set_volume(percentage):
    """
    Ajusta o volume do sistema para um percentual específico (0 a 100).
    """
    # Garante que o valor digitado fique estritamente entre 0 e 100
    percentage = max(0, min(100, percentage))
    os_name = platform.system()

    try:
        if os_name == 'Windows':
            from pycaw.pycaw import AudioUtilities
            
            # Obtém o dispositivo de áudio principal (alto-falantes)
            device = AudioUtilities.GetSpeakers()
            
            # Na nova API do pycaw, acessamos a interface de volume diretamente
            volume = device.EndpointVolume
            
            # Ajusta o volume (aceita escala de 0.0 a 1.0)
            volume.SetMasterVolumeLevelScalar(percentage / 100.0, None)
            print(f"Volume do Windows ajustado para {percentage}%")

        elif os_name == 'Darwin': # macOS
            comando = f"set volume output volume {percentage}"
            subprocess.run(["osascript", "-e", comando], check=True)
            print(f"Volume do macOS ajustado para {percentage}%")

        elif os_name == 'Linux':
            comando = ["amixer", "-D", "pulse", "sset", "Master", f"{percentage}%"]
            subprocess.run(comando, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL, check=True)
            print(f"Volume do Linux ajustado para {percentage}%")

        else:
            print(f"Sistema operacional '{os_name}' não suportado por este script.")

    except Exception as e:
        print(f"Ocorreu um erro ao tentar ajustar o volume no {os_name}: {e}")

# --- Testando o script ---
if __name__ == "__main__":
    set_volume(12)
```

```python
import random

def pick_one_by_one(items):
    # Copy and shuffle to protect the original list
    pool = list(items)
    random.shuffle(pool)
    for item in pool:
        yield item

my_list = ['apple', 'banana', 'cherry', 'date']
picker = pick_one_by_one(my_list)

# Get the next random item whenever you need it
print(next(picker))  # e.g., 'cherry'
print(next(picker))  # e.g., 'banana'
```
