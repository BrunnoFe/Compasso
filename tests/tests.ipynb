```python
import tkinter as tk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure
import random

# Initial setup
root = tk.Tk()
fig = Figure(figsize=(5, 4))
ax = fig.add_subplot(111)
line, = ax.plot([], [], 'r-')
canvas = FigureCanvasTkAgg(fig, master=root)
canvas.get_tk_widget().pack()

x_data, y_data = [], []

def update_plot():
    # Append data, update line object, and redraw
    y_data.append(random.randint(0, 100))
    x_data.append(len(y_data))
    line.set_data(x_data, y_data)
    ax.relim()
    ax.autoscale_view()
    canvas.draw()
    root.after(100, update_plot) # Repeat every 100ms

update_plot()
root.mainloop()
```

```python
import os
import pandas as pd

# Define your target folder and output file path
folder_path = r"C:\Users\shaulimdoplim\Documents\GitHub\Compasso\musicas"
output_excel = "file_list.xlsx"

# 1. Collect all file names and paths
file_data = []
for root, dirs, files in os.walk(folder_path):
    for file in files:
        full_path = os.path.join(root, file)
        file_data.append({
            "File Name": file,
            "Folder Location": root,
            "Full Path": full_path
        })

# 2. Convert to a DataFrame and export to Excel
df = pd.DataFrame(file_data)
df.to_excel(output_excel, index=False)

print(f"Successfully saved {len(df)} files to '{output_excel}'.")
```

```python
import os
from moviepy.editor import VideoFileClip

input_folder = r"C:\Users\shaulimdoplim\Videos\4K Video Downloader+\Para Bruno"
output_folder = r"C:\Users\shaulimdoplim\Videos\4K Video Downloader+\mp3"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Loop and convert all MP4 files
for filename in os.listdir(input_folder):
    if filename.endswith(".mp4"):
        video_path = os.path.join(input_folder, filename)
        
        # Change the extension for the output name
        audio_filename = filename.replace(".mp4", ".mp3")
        audio_path = os.path.join(output_folder, audio_filename)
        
        print(f"Converting: {filename}...")
        video = VideoFileClip(video_path)
        video.audio.write_audiofile(audio_path)
        video.close()

print("All files converted successfully!")
```
